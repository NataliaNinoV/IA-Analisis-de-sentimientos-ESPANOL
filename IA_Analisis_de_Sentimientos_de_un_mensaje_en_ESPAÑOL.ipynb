```python
import random
frases_seleccionadas2 = random.sample(frases2, 5)

for n in frases_seleccionadas2:
    print(n)
```

```python
from textblob import TextBlob
from googletrans import Translator

translator = Translator()
texto = frases2 # todas las frases en español
comentarios = [] #Aqui se almacenará la traducción de todas las frases
contar = 0

# Traducir cada frase del español al ingles
for cont in range(len(texto)):
    traduccion = translator.translate(texto[contar], src='es', dest='en') # Traducir cada palabra
    comentarios.append(traduccion.text + "") #espacio entre palabras
    #print(comentarios)
    contar = contar + 1
```

```python
positive_feedbacks = []
negative_feedbacks = []
neutral_feedbacks = []
traduccion =[]


# evaluacion de frases y puntaje respecto a su valor
for i in comentarios:
    puntacion = TextBlob(i).sentiment.polarity
    print(i + "  :         " + str(puntacion))

  # almacenar cada resultado en un vector
    if puntacion > 0:
        positive_feedbacks.append(i)
    elif puntacion < -0:
        negative_feedbacks.append(i)
    else:
        neutral_feedbacks.append(i)
```

```python
print("Comentarios positivos:")
for feedback in positive_feedbacks:
    print("- " + feedback)

print("Comentarios negativos:")
for feedback in negative_feedbacks:
    print("- " + feedback)

print("Comentarios neutrales:")
for feedback in neutral_feedbacks:
    print("- " + feedback)
```
